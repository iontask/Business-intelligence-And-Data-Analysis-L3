```python
# Initialisation du client Pub/Sub
project_id = 'ton-id-de-projet'  # Remplace par ton ID de projet
topic_id = 'ton-topic-id'  # Remplace par ton ID de topic

publisher = pubsub_v1.PublisherClient()
topic_path = publisher.topic_path(project_id, topic_id)

# Fonction pour envoyer des données vers Pub/Sub
def publish_data_to_pubsub(data):
    data_str = str(data)
    data_bytes = data_str.encode('utf-8')  # Pub/Sub accepte les données sous forme de bytes
    future = publisher.publish(topic_path, data_bytes)
    print(f"Message publié: {data_str}")

# Exemple de données à envoyer
data_sample = {'timestamp': '2024-11-21T12:00:00', 'temperature': 22.5, 'humidity': 60}
publish_data_to_pubsub(data_sample)
```

```python
import time

# Function to send the data to Pub/Sub continuously
def send_data_stream(max_iterations=None):  # Added max_iterations parameter
    # Assuming 'ts' is the actual column name for timestamp in your DataFrame
    timestamp_column_name = 'ts'

    # Get the actual column names from your DataFrame
    column_names = data.columns.tolist()
    print(f"Column names in your DataFrame: {column_names}")  # Print the column names

    # Determine the correct column name for temperature
    temperature_column_name = None
    for name in column_names:
        if "temp" in name.lower():  # Check for variations of "temp"
            temperature_column_name = name
            break

    if temperature_column_name is None:
        raise KeyError("Temperature column not found in DataFrame. Please check your data.")

    iteration_count = 0  # Initialize iteration counter
    for index, row in data.iterrows():
        row_data = {
            'timestamp': row[timestamp_column_name],  # Access using the correct column name
            'temperature': row[temperature_column_name],  # Access using the determined temperature column name
            'humidity': row['humidity']
            # Removed 'pressure': row['pressure'] because it's causing the KeyError
        }

        # Check if 'pressure' column exists before adding it to row_data
        if 'pressure' in data.columns:
            row_data['pressure'] = row['pressure']

        # Send each row as a message to Pub/Sub
        publish_data_to_pubsub(row_data)

        # Pause for 1 second to simulate an IoT data stream
        time.sleep(1)

        iteration_count += 1  # Increment iteration counter

        # Check if max_iterations is reached and stop if so
        if max_iterations is not None and iteration_count >= max_iterations:
            print(f"Simulation stopped after {max_iterations} iterations.")
            break  # Exit the loop

# Launch the data stream with a maximum of 10 iterations
send_data_stream(max_iterations=10)
```

```python
# Filtrer les anomalies (exemple : température > 30°C)
# Get the correct temperature column name
temperature_column_name = None
for name in data.columns:
    if "temp" in name.lower():  # Check for variations of "temp"
        temperature_column_name = name
        break

if temperature_column_name is None:
    raise KeyError("Temperature column not found in DataFrame. Please check your data.")

# Use the correct temperature column name for filtering
anomalies = data[data[temperature_column_name] > 30]

# Sauvegarder les anomalies dans un fichier CSV local
anomalies.to_csv('anomalies.csv', index=False)
print("Fichier 'anomalies.csv' sauvegardé localement.")
```

```python
# Afficher les anomalies détectées
print("Rapport des anomalies détectées :")

# Get the correct column names if they differ from 'timestamp' and 'temperature'
timestamp_column = 'ts'  # Replace with the actual timestamp column name in your DataFrame
temperature_column = None  # Initialize to None
for name in anomalies.columns:
    if "temp" in name.lower():  # Check for variations of "temp"
        temperature_column = name
        break

if temperature_column is None:
    raise KeyError("Temperature column not found in DataFrame. Please check your data.")

# Print the anomalies using the correct column names
print(anomalies[[timestamp_column, temperature_column]])

# Exporter en un fichier texte
with open('anomaly_report.txt', 'w') as f:
    for _, row in anomalies.iterrows():
        # Use the correct column names for timestamp and temperature
        f.write(f"Anomalie détectée - Timestamp : {row[timestamp_column]}, Température : {row[temperature_column]}°C\n")
print("Rapport 'anomaly_report.txt' généré.")
```
