```python
import csv
import random

# Générer des données aléatoires
noms = ["Jean", "Sophie", "Amina", "Mohamed", "Fatima", "Paul", "Lucie"]
prenoms = ["Dupont", "Leroy", "Bouh", "Khalid", "Ben", "Mouline", "Kanté"]
filieres = ["Informatique", "IA", "Gestion", "Droit", "Finance"]

# Générer un fichier CSV
with open('etudiants.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Nom", "Prenom", "Age", "Filiere"])  # En-tête
    for _ in range(50):  # 50 étudiants
        writer.writerow([
            random.choice(noms),
            random.choice(prenoms),
            random.randint(18, 30),
            random.choice(filieres)
        ])

print("Fichier CSV 'etudiants.csv' créé avec succès.")
```

```python
from pyspark.sql import SparkSession

# Créer une session Spark
spark = SparkSession.builder.appName("ETL_Students").getOrCreate()

# Charger le fichier CSV
df = spark.read.csv('etudiants.csv', header=True, inferSchema=True)

# Afficher les données
df.show()
```

```python
from pyspark.sql.functions import col, upper, avg

# Convertir les noms en majuscules
df_transformed = df.withColumn("Nom", upper(col("Nom")))

# Filtrer les étudiants ayant plus de 20 ans
df_filtered = df_transformed.filter(col("Age") > 20)

# Calculer la moyenne d'âge par filière
df_stats = df_filtered.groupBy("Filiere").agg(avg("Age").alias("Moyenne_Age"))

# Afficher les résultats
df_stats.show()
```

```python
# Sauvegarder dans un nouveau fichier CSV
df_stats.write.csv('resultats_etudiants.csv', header=True)
print("Données sauvegardées dans 'resultats_etudiants.csv'.")
```

```python
from confluent_kafka import Producer
import json
import csv
from collections import defaultdict

# Configuration du producteur Kafka
bootstrap_servers = 'kafka-18536ef0-isy03436-05ce.g.aivencloud.com:14686'
ssl_cafile = '/content/ca.pem'
ssl_certfile = '/content/service.cert'
ssl_keyfile = '/content/service.key'

conf = {
    'bootstrap.servers': bootstrap_servers,
    'security.protocol': 'SSL',
    'ssl.ca.location': ssl_cafile,
    'ssl.certificate.location': ssl_certfile,
    'ssl.key.location': ssl_keyfile
}

# Création d'un producteur Kafka
producer = Producer(conf)

# Exemple de données brutes
csv_data = "/content/etudiants.csv"

# ** Transformations **

# 1. Convertir les noms en majuscules et filtrer les étudiants de plus de 20 ans
# The issue was caused by incorrect indentation and an extra closing bracket.
# The following code fixes those issues.
# It also assumes you want to read data from the csv file.
# Make sure you have the 'name', 'Age', and 'filiere' columns in your CSV.

with open(csv_data, mode='r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    transformed_data = [
        {
            "name": row["Nom"].upper(),
            "age": int(row["Age"]),  # Assuming Age is an integer
            "filiere": row["Filiere"]
        }
        for row in reader if int(row["Age"]) > 20
    ]


# 2. Calculer la moyenne d'âge par filière
stats = defaultdict(list)
for student in transformed_data:
    stats[student["filiere"]].append(student["age"])

average_age_by_filiere = {
    filiere: sum(ages) / len(ages) for filiere, ages in stats.items()
}

# Affichage des statistiques
print("\nStatistiques : Moyenne d'âge par filière")
for filiere, avg_age in average_age_by_filiere.items():
    print(f"Filière : {filiere}, Âge moyen : {avg_age:.2f}")

# 3. Sauvegarde des données transformées dans un fichier CSV
output_file = "transformed_etudiants.csv"
with open(output_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=["name", "age", "filiere"])
    writer.writeheader()
    writer.writerows(transformed_data)

print(f"\nDonnées transformées sauvegardées dans le fichier : {output_file}")

# ** Envoi des données transformées à Kafka **
topic_name = 'etudiants'

def delivery_report(err, msg):
    if err is not None:
        print(f"Delivery failed for record {msg.key()}: {err}")
    else:
        print(f"Record successfully produced to {msg.topic()} [{msg.partition()}] @ offset {msg.offset()}")

# Envoi des données transformées au topic Kafka
for student in transformed_data:
    producer.produce(topic_name, key=str(student["name"]), value=json.dumps(student), callback=delivery_report)

# Vidange des messages
producer.flush()

print("\nTous les messages ont été envoyés à Kafka.")













```
